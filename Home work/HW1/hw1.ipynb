##### Apply different normalization methods to the selected features

```python
from sklearn.preprocessing import StandardScaler,MinMaxScaler,RobustScaler,PowerTransformer

# StandardScaler
standard_scaler = StandardScaler()
standard_scaler_data = standard_scaler.fit_transform(X)

# MinMaxScaler
minmax_scaler = MinMaxScaler()
minmax_scaler_data = minmax_scaler.fit_transform(X)

# RobustScaler
robust_scaler = RobustScaler()
robust_scaler_data = robust_scaler.fit_transform(X)

# PowerTransformer
power_transformer = PowerTransformer()
power_transformer_data = power_transformer.fit_transform(X)



```
